```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('books.csv')
print("Загружены данные:")
print(df.head())
print(f"\nСтруктура данных: {df.shape}")
print(f"Столбцы: {df.columns.tolist()}")

def plot_books_2023(df):
    df_2023 = df[df['Год'] == 2023]
    months_order = ['Январь', 'Февраль', 'Март', 'Апрель', 'Май', 'Июнь',
                    'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь']

    plt.subplot(2, 1, 1)
    plt.plot(df_2023['Месяц'], df_2023['Книги'],
             marker='o', color='red', label='Прочитано книг')

    plt.title('Прочитанные книги 2023')
    plt.xlabel('Месяцы')
    plt.ylabel('Количество книг')
    plt.grid(True)
    plt.legend()

    plt.xticks(df_2023['Месяц'])

def plot_pages_2024(df):
    df_2024 = df[df['Год'] == 2024]
    plt.subplot(2, 1, 2)
    plt.bar(df_2024['Месяц'], df_2024['Страницы'],
            color='green', label='Страницы')

    plt.title('Прочитанные страницы 2024')
    plt.xlabel('Месяцы')
    plt.ylabel('Количество страниц')
    plt.grid(True)
    plt.legend()

    plt.xticks(df_2024['Месяц'])

def create_combined_plot(df):
    plt.figure(figsize=(12, 8))
    plot_books_2023(df)
    plot_pages_2024(df)
    plt.tight_layout()
    plt.show()

create_combined_plot(df)
```
